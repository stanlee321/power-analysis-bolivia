# SIN ANALYSIS

```python
import pandapower as pp
import pandas as pd
```

```python
net = pp.create_empty_network()
```

### Lines

```python
excel_path = './data/DATOS_SIN_TOTAL.xls'

hv_lines = pd.read_excel(excel_path,sheet_name=1, header = 6, skyprows = 5)
hv_lines.head()
```

```python
hv_lines.columns
```

```python
### Create standar types
"""
Avaible columns

Index(['AREA', 'PROPIETARIO', 'EN EL STI', 'TRAMO', 'TIPO', 'CODIGO',
       'Nº CIRCUITOS', 'Un (kV)', 'L (km)', 'Pn (MW)', 'NODO 1',
       'INTERRUPTOR 1', 'NODO 2', 'INTERRUPTOR 2', 'In (A)', 'Sbase (MVA)',
       'R1 (ohm)', 'X1 (ohm)', 'C1 (F)', 'R0 (ohm)', 'X0 (ohm)', 'C0 (F)',
       'Zc (ohm)', 'Sn (MVA)', 'SOBRECARGA (MVA)', 'R1 (%)', 'X1 (%)',
       'Qvacio (MVAr)'],
      dtype='object')
"""
for index, row in hv_lines.iterrows():
    n_type = {
        "Un[kV]": row['Un (kV)'],
        "L[km]":row['L (km)'],
        "Pn[MW]":row['Pn (MW)'],
        "In[A]": row['In (A)'],
        "Sbase[MVA]":row['Sbase (MVA)'],
        "R1[ohm]": row['R1 (ohm)'],
        "X1[ohm]": row['X1 (ohm)'],
        "C1[F]": row['C1 (F)'],
        "R0[ohm]": row['R0 (ohm)'],
        "X0[ohm]": row['X0 (ohm)'],
        "C0[ohm]": row['C0 (F)'],
        "Zc[ohm]": row['Zc (ohm)'],
        "Sn[MVA]": row['Sn (MVA)'],
        "SOBRECARGA[MVA]": row['SOBRECARGA (MVA)'],
        "R1%": row['R1 (%)'],
        "X1%": row['X1 (%)'],
        "Qvacio[MVAr]": row['Qvacio (MVAr)'],
        "Tipo": row['TIPO'],
        "INTERRUPTOR 1": row['INTERRUPTOR 1'],
        "INTERRUPTOR 2": row['INTERRUPTOR 2'],
        "r_ohm_per_km": row['R1 (ohm)']/row['L (km)'], 
        "x_ohm_per_km": row['X1 (ohm)']/row['L (km)'], 
        "c_nf_per_km": row['C1 (F)']/row['L (km)'],
        "max_i_ka": row['In (A)']/1000,
        "type": "cs",
        "h_per_km": 0.01
    }
    pp.create_std_type(net, name=row['CODIGO'], data=n_type, element="line")
pp.available_std_types(net, element="line")
```

### Nodes

```python
hv_nodes = pd.read_excel(excel_path, sheet_name=0, header = 7, skyprows = 0)
hv_nodes = hv_nodes.drop(hv_nodes.index[0])
hv_nodes.head()
```

```python
hv_nodes.columns
```

### Create buses

```python
# Create buses from sheet 0
for index, row in hv_nodes.iterrows():
    
    pp.create_bus(net, name=row["CODIGO"], vn_kv=row["NOMINAL"], type='n', zone=row["AREA"])

    net.bus
```

### Create lines

```python
# create lines
for _, hv_line in hv_lines.iterrows():
    try:
        from_bus = pp.get_element_index(net, "bus", hv_line['NODO 1'])
        to_bus = pp.get_element_index(net, "bus", hv_line['NODO 2'])
        pp.create_line(net, from_bus, to_bus, length_km=hv_line["L (km)"],
                   std_type=hv_line['CODIGO'], 
                   name= hv_line["TRAMO"], 
                   parallel=hv_line["Nº CIRCUITOS"])
    except Exception as e:
        print(e)
# show line table
net.line
```

```python
pp.parameter_from_std_type(net, parameter="Sbase[MVA]")
pp.parameter_from_std_type(net, parameter="Pn[MW]")

net.line
```

```python
### Trasformers

hv_trafos = pd.read_excel(excel_path,sheet_name=3, header = 6, skyprows = 5)
hv_trafos.head()
```

```python
hv_trafos.columns
```

```python
#### Create transformers
def from_conn2angle(conn):
    if '11' in conn:
        return 30
    elif '0' in conn:
        return 0
    elif '5' in conn:
        return 150
    else:
        return 60

for _, row in hv_trafos.iterrows():
    try:

        hv_bus = pp.get_element_index(net, "bus",  row['Nodo 1'])
        lv_bus = pp.get_element_index(net, "bus",  row['Nodo 2'])

        pp.create_transformer_from_parameters(net, hv_bus,
                                              lv_bus, 
                                              sn_kva=row['Capacidad Nominal (MVA)'], 
                                              vn_hv_kv=row['U1n \n(kV)'], 
                                              vn_lv_kv=row['U2n \n(kV)'], 
                                              vscr_percent=row['R1 (pu)'],
                                              vsc_percent=row['X1 (pu)'], 
                                              pfe_kw=row['ΔPf (kW)'],
                                              i0_percent=0, 
                                              tp_pos=row['Posicion Tap Actual (pu)'],
                                              tp_max=row['Posicion Tap Max (pu)'],
                                              tp_min=row['Posicion Tap Mín (pu)'],
                                              tp_st_percent=row['Paso entre taps (pu)'],
                                              shift_degree=from_conn2angle(row['Grupo de Conexión']), 
                                              name=row['NOMBRE'])
    except Exception as e:
        print(e)

net.trafo # show trafo table
```
